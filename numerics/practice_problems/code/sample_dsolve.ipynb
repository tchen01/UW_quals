```python
import numpy as np
import matplotlib.pyplot as plt
import sympy
from sympy import *
init_printing(use_unicode=True)
```

## Solve System

Set up symbolic variables

```python
u = Function('u')
x = symbols('x',real=True)
k = symbols('k',positive=True)
```

Solve $u'(x) + k^2 u(x) = 0$

```python
dsol = dsolve(Derivative(u(x),x,2)+k**2*u(x),u(x))
u_sol = dsol.rhs
```

solve for constants using boundary conditions $u(0)=1$ and $u'(1) = iku(1)$

```python
csts=solve([u_sol.subs(x,0)-1,Derivative(u_sol,x).subs(x,1).doit()-I*k*u_sol.subs(x,1)])
```

subsitute constants back in, and simplify expression

```python
u_sol=u_sol.subs(csts[0]).expand(complex=True).simplify()
```

## Plot Solution

convert symbolic expression to expression which can be evaluted in numpy with $k=75$

```python
u_sol_np=lambdify(x,u_sol.subs(k,75),'numpy')
```

evaluate solution on mesh
